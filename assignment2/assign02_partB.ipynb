### Part B: RESP savings table

In the first assignment you explored writing the Python code for an arithmetic expression involving the future value of money.  Given the age of the child at the start of the RESP, the monthly amount to be saved, and the annual rate of return of the investment, your program calculated the total amount of the RESP when the child reaches the age of 18.

In this part you will re-use that calculation, but now print a table showing the value of the RESP at the end of each year for which funds were saved in the RESP. That is, after asking for:
* age of child at the start of the RESP
* monthly contribution amount
* annual rate of return of investment in percent
your program will print a table for each year from the start of the RESP up to and including 18. 

Using the example from assignment one (age at start: 3; monthly saving: \$150; annual rate of return: 10%), the following table would be generated by your program.
```
age   RESP value
 4      1884.84
 5      3967.04
 6      6267.27
 7      8808.37
 8     11615.56
 9     14716.70
10     18142.56
11     21927.16
12     26108.06
13     30726.75
14     35829.07

```

Write a program that produce this table. The program must ask the user to input the following (**and in this order**):

1. The age of the child at the start of the RESP (i.e., assume the RESP starts on the child's birthday)
1. The amount to be invested each month
1. The annual rate of return for the investment, in percent.

and after this print the table with numbers aligned as shown in the example (i.e., integers aligned as shown, and floating-point number as shown). When writing your program, you will find it easier to simply recalculate the RESP value for each year of the child's youth.



````python
# Put in solution to Part B here.


#ask the user to input the start age
start_age = int(input("Please input the age of the child at the start of the RESP:"))

#ask the user to input the monthly amount
M = int(input("Please input amount to be invested each month:"))

#ask the user to input the annual rate of return for the investment, in percent.
rate = int(input("Please input the annual rate of return for the investment, in percent:"))

i = rate/100/12

print("{0}\t{1}".format("age","RESP value"))
for age in range(start_age+1,19):
    n=(age-start_age)*12
    #calculate the amount of money that will be in the RESP
    RESP = M*(((1+i)**n-1)/i)
    #print the result
    print("{0:^3}\t".format(str(age)), end="")
    print(format(RESP,"9.2f"))
````
